```python
import numpy as np
import pandas as pd
import urllib.request
from io import StringIO
```

```python
url = "https://raw.githubusercontent.com/facebook/prophet/main/examples/example_wp_log_peyton_manning.csv"
```

```python
# ref: https://yaju3d.hatenablog.jp/entry/2018/06/30/122754
def read_csv_from_url(url: str) -> pd.DataFrame:
    res = urllib.request.urlopen(url)
    res = res.read().decode("utf-8")
    df = pd.read_csv(StringIO(res))
    return df
```

```python
df = read_csv_from_url(url)
```

```python
df.head(5)
```

```python
df["ds"] = pd.to_datetime(df["ds"])
```

```python
np.random.seed(2022)
feature_1 = pd.DataFrame({
    "target_ds": df["ds"],
    "flag_1": np.random.binomial(1, 0.25, len(df))
})
feature_2 = pd.DataFrame({
    "target_ds": df["ds"],
    "flag_2": np.random.binomial(1, 0.1, len(df))
})
flag_df_list = [feature_1, feature_2]
```

```python
for _df_flag in flag_df_list:
    df = pd.merge(
        left=df,
        right=_df_flag,
        how="left",
        left_on="ds",
        right_on="target_ds"
    )
df.head()
```
